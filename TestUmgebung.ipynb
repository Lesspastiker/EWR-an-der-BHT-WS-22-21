## Plot mit Sinusfunktion

```python
import math as m 

B = Stripmethod(lambda x: m.sin(x) , -10, 10, 100) 
B.plotObersumme()
```

## Plot mit Python Expression

```python
# lambda_parse als interpreter des python Strings
from Stripmethod import lambda_parse

S = Stripmethod(lambda_parse("2*x**2"), -2, 2, 10)
S.plotOberUnter()
```

## Kompliziertes Beispiel

```python
from Stripmethod import lambda_parse
from Stripmethod import Stripmethod

A = Stripmethod(lambda_parse("3*x**3-4*x"), -5, 5, 100)
A.plotOberUnter()
```
